```python
import json
import requests
import pandas as pd
from datetime import datetime

# insert your API key here
API_KEY = ''# key



# make API request
resBtc = requests.get(

        'https://api.glassnode.com/v1/metrics/derivatives/options_atm_implied_volatility_all',
        params={'a': 'BTC', 'i': '1h', 'api_key': API_KEY, 's':1640966400,'u': 1688713200})

# # convert to pandas dataframe
dfBtc = pd.read_json(resBtc.text, convert_dates=['t'])
```

```python
iv = []
for i in dfBtc.index:
    iv.append(dfBtc.iloc[i]['o']['1m'])

dfBtc['iv'] = iv

dfBtc_ = dfBtc.set_index('t')
```
